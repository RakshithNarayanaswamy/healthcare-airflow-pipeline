```python
from pyspark.sql.functions import col, when

vitals_silver = spark.readStream.table(
    "healthcare_analytics.silver_vitals_stream"
)
```

```python
vitals_gold = (
    vitals_silver
    .withColumn(
        "risk_level",
        when(
            (col("heart_rate") > 120) | 
            (col("oxygen_level") < 90),
            "HIGH"
        )
        .when(
            (col("heart_rate") > 100),
            "MEDIUM"
        )
        .otherwise("NORMAL")
    )
)
```

```python
(
    vitals_gold.writeStream
    .format("delta")
    .option(
        "checkpointLocation",
        "/Volumes/workspace/healthcare_analytics/datastore/checkpoints/vitals_gold"
    )
    .outputMode("append")
    .trigger(availableNow=True)
    .toTable("healthcare_analytics.gold_vitals_risk")
)
```
