```python
import time
import random
from datetime import datetime
import pandas as pd
```

```python
STREAM_PATH = "/Volumes/workspace/healthcare_analytics/dataextract/vitals_stream/"
```

```python
def generate_vitals_batch(batch_size=50):
    data = []

    for _ in range(batch_size):
        data.append({
            "patient_id": random.randint(1, 500),
            "heart_rate": random.randint(55, 130),
            "systolic_bp": random.randint(90, 180),
            "oxygen_level": random.randint(85, 100),
            "event_time": datetime.utcnow()
        })

    return pd.DataFrame(data)
```

```python
i = 0

while i < 20:   
    df = generate_vitals_batch()

    output_path = f"{STREAM_PATH}/batch_{i}.csv"
    df.to_csv(output_path, index=False)

    print(f"Wrote {output_path}")
    time.sleep(5)  
    i += 1
```
